```python
a = np.linspace(0,10,11)
```

```python
b = a**4
```

```python
a
```

```python
b
```

```python
x = np.arange(0,10)
```

```python
y = 2*x
```

```python
x
```

```python
y
```

```python
fig = plt.figure()

axes1 = fig.add_axes([0,0,1,1])
axes1.set_xlabel('A')
axes1.set_ylabel('B')
axes1.set_xlim(0,8)
axes1.set_ylim(0,8000)
axes1.set_title('Power of 4'); 
axes1.plot(a,b);

axes2 = fig.add_axes([0.2,0.5,0.25,0.25])
axes2.set_xlabel('A')
axes2.set_ylabel('B')
axes2.set_xlim(1,2)
axes2.set_ylim(0,50)
axes2.set_title('Zoomed In');
axes2.plot(x,y);
```

```python
fig = plt.figure(dpi=400, figsize=(10,3))
axes1 = fig.add_axes([0,0,1,1])
axes1.set_xlabel('A')
axes1.set_ylabel('B')
axes1.set_title('Power of 4');
axes1.plot(a,b);
fig.savefig('new_figure.png', bbox_inches = 'tight')
```

```python
fig,axes = plt.subplots(nrows=1,ncols=2)
axes[0].plot(x,y)
axes[1].plot(a,b)
```

```python
fig,axes = plt.subplots(nrows=2,ncols=2)
axes[0][0].plot(x,y)
axes[1][0].plot(a,b)
```

```python
type(axes)
```

```python
fig,axes = plt.subplots(nrows=3,ncols=1)

for ax in axes:
  ax.plot(x,y)

plt.tight_layout()
```

```python
fig,axes = plt.subplots(nrows=3,ncols=1)

for ax in axes:
  ax.plot(x,y)

fig.subplots_adjust(wspace=1,hspace=0.5)
```

```python
fig,axes = plt.subplots(nrows=2,ncols=2,dpi=400,figsize=(10,3))
axes[0][0].plot(x,y)
axes[0][0].set_ylabel('Y LABEL')
axes[0][1].plot(a,b)
axes[1][0].plot(a,b)
axes[1][1].plot(a,b)

fig.suptitle('FIGURE')
fig.subplots_adjust(wspace=1,hspace=1)
#plt.tight_layout()
fig.set_figwidth(10)
fig.savefig('new_subplots.png',bbox_inches='tight')
```

```python
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
ax.plot(x,x,label='X vs X')
ax.plot(x,x**2,label="X vs X^2")

ax.legend()
```

```python
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
ax.plot(x,x,color='#034212')
```

```python
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
ax.plot(x,x,color='#034212',label='X vs X')
ax.plot(x,x+1,color='#e34009',label='X vs X+1')

ax.legend()
```

```python
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
ax.plot(x,x,color='#034212')
ax.plot(x,x+1,color='#e34009', lw=10)
```

```python
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
ax.plot(x,x,color='#034212',ls='-.')
ax.plot(x,x+1,color='#e34009', ls='--',lw=5)
```

```python
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
ax.plot(x,x,color='#034212',ls='-.')
ax.plot(x,x+1,color='#e34009', ls='--',marker='o')
```

```python
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
ax.plot(x,x,color='#034212',ls='-.')
ax.plot(x,x+1,color='#e34009', ls='--',marker='o',markersize=20)
```
